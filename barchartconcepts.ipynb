# bar chart

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

a=[1,2,3,4]
b=[1,4,9,16]
plt.bar(a,b,color="black")
plt.title("Value VS square barchart",color="blue")
plt.xlabel("values")
plt.ylabel("squares")
plt.show()
```

```python
children=[10,20,30,50]
colors=["red","blue","green","pink"]
plt.bar(colors,children)
plt.xlabel("colors")
plt.ylabel("children")
plt.show()
```

```python
cities=['delhi','mumbai','banglore','hyderabad','chennai']
population=[2004783,1997863,1876907,2816792,1872974]
plt.bar(cities,population)
plt.xlabel("cities")
plt.ylabel("population")
plt.show()
```

```python
df=pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"],df["2015"],width=0.5)
plt.xlabel("batsman-->",color="blue")
plt.ylabel("runs scored-->",color="blue")
plt.show()
```

```python
import numpy as np
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

```

# plt.xticks function is used for display x axis values for example-->

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df["batsman"])
plt.show()
```

# horizontal bar chart

<h3> plt.barh () function is used to create a horizontal barchart and its look like this

```python
df=pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.barh(df['batsman'],df['2015'])
plt.xlabel("runs scored")
plt.ylabel("batsman")
plt.show()
```

# problem arise in this 

<h3> categories ka naam bda ho toh wo ek dusro ko overlap krti hai jisey readability zero ho jati h x-axis par jaise ki 

```python
children=[10,20,30,50]
colors=["redredredred","bluebuleblueblue","greengreengreengreen","pinkpinkpinkpink"]
plt.bar(colors,children)
plt.xlabel("colors")
plt.ylabel("children")
plt.show()
```

# toh is problem ko hum solve krte h by plt.xticks(rotation="vertical") karke

```python
children=[10,20,30,50]
colors=["redredredred","bluebuleblueblue","greengreengreengreen","pinkpinkpinkpink"]
plt.bar(colors,children)
plt.xlabel("colors")
plt.ylabel("children")
plt.xticks(rotation="vertical") 
plt.show()
```
