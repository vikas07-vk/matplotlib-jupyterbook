```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#from simple data
```

```python
data=[23,45,100,20,44]
plt.pie(data)
plt.show()
```

```python
# adding labels 
data=[23,45,100,20,44]
subjects=['AI','ML','OOPS','C++','JAVA']
plt.pie(data, labels=subjects)
plt.show()
```

```python
# showing percentage on chart we use autopct function
```

```python
data=[23,45,100,20,44]
subjects=['AI','ML','OOPS','C++','JAVA']
plt.pie(data, labels=subjects, autopct='%0.1f%%')
plt.show()
```
